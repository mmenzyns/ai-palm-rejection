```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np

```

```python
neurons = [1, 2, 4, 8, 16, 32, 64, 128, 256]
accuracies = [0.6305903792381287, 0.635062575340271, 0.847048282623291, 0.8434704542160034, 0.9025045037269592, 0.915026843547821, 0.9132379293441772, 0.9078711867332458, 0.9212880730628967]

sns.lineplot(x=neurons, y=np.array(accuracies)*100)
```

```python
data = {}
f = open('logs/cnn-filters-1kernel')

for line in f:
    if line == '':
        continue
    tokens = line.split(' ')
    if len(tokens) < 3:
        continue

    filter_num = int(tokens[0])
    if not filter_num in data:
        data[filter_num] = []

    accuracy = float(tokens[2].strip('%\n'))
    data[filter_num].append(accuracy)
f.close()

x = []
y = []
for key, item in data.items():
    x.append(key)
    y.append(np.mean(item))

sns.lineplot(x, y)
plt.suptitle("Accuracy per number of filters in convolutional hidden layer with 1 kernel")
list(zip(x,np.round(y, 1)))
```
