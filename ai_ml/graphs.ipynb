```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np

```

```python
# CNN 0.2 dropout 2 kernels
data = {}
f = open('logs/cnn-filters-2kernels-dropout')

for line in f:
    if line == '':
        continue
    tokens = line.split(' ')
    if len(tokens) < 3:
        continue

    filter_num = int(tokens[0])
    if not filter_num in data:
        data[filter_num] = []

    accuracy = float(tokens[2].strip('%\n'))
    data[filter_num].append(accuracy)
f.close()

x = []
y = []
for key, item in data.items():
    x.append(key)
    y.append(np.mean(item))

sns.lineplot(x, y)
plt.xlabel("Filters")
plt.ylabel("Accuracy in percents")
plt.savefig("figures/cnn-filters.pdf")
# plt.suptitle("Accuracy per number of filters in convolutional hidden layer with 2 kernels and dropout 0.2")
# list(zip(x,np.round(y, 1)))
```
